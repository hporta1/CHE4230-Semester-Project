```python
#Method 1 for DR ISOMAP
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_excel(r"./../Data/RawData.xlsx")
```

```python
#scaling the data

scale = StandardScaler().fit(df)

df_scaled = scale.transform(df)
```

```python
from sklearn.manifold import Isomap

isomap = Isomap(n_components=2)
isomap_result = isomap.fit_transform(df_scaled)
isomap_result.shape

#checking to make sure the shape is right
```

```python
import matplotlib.pyplot as plt

plt.scatter(isomap_result[:,0], isomap_result[:,1])
plt.title("ISOMAP", fontsize=24)
```
